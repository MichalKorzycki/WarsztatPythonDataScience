## Funkcje

```python
def dodaj_2(x):
    wynik = x + 2
    return wynik
```

```python
dodaj_2(5)
```

```python
def is_odd(x):
    print ("*" * x)
    return (x % 2) == 1
```

```python
is_odd(5)
```

```python
is_odd(8)
```

```python
def slownie(n):
    jednosci = { 0: "zero", 1: "jeden", 2: "dwa", 3: "trzy", 4: "cztery", 5: "pięć", 6: "sześć", 7: "siedem", 8: "osiem", 9: "dziewięć"}
    return jednosci[n]

slownie(6)
```
