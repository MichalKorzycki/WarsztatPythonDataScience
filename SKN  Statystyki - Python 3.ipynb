## Zadanie 1

Stworz funkcję która zwróci drugi najmniejszy element listy

druginajmniejszy([1, 2, 3, 4, 6])

2

```python
def druginajmniejszy(lista):
    if len(lista) < 2:
        return lista[0]
    if lista[0] > lista[1]:
        pierwszy, drugi = lista[1], lista[0]
    else:
        pierwszy, drugi = lista[0], lista[1]
    for x in lista[2:]:
        #print x, pierwszy, drugi
        if x < drugi:
            if x < pierwszy:
                pierwszy, drugi = x, pierwszy
            else:
                drugi = x
    return drugi
```

```python
druginajmniejszy([2,1,3,4,5])
```

```python
def usun_min(lista):
    for i in range(len(lista)):
        if lista[i]==min(lista):
            del lista[i]
```

```python
l = [4, 1, 2, 3]
usun_min(l)
l
```

```python
def usun_min(lista):
    do_usuniecia = set()
    for i in range(len(lista)):
        if lista[i]==min(lista):
            do_usuniecia.add(i)
    for i in do_usuniecia:
        del lista[i]
```

```python
l = [4, 1, 2, 3]
usun_min(l)
l
```

## Zadanie 3

Stworz funkcję która usunie "wyplaszczy" zagnieżdzone listy

l = [4, 1, [ 2, [7] ], 3]

wyplaszcz(l)

l

[4, 1, 2, 7, 3]

```python
def wyplaszcz(lista):
    ret = []
    for x in lista:
        if type(x)==type([]):
            for y in wyplaszcz(x):
                ret.append(y)
        else:
            ret.append(x)
    return ret
```

```python
l = [4, 1, [ 2, [7] ], 3]
l = wyplaszcz(l)
l
```

```python
def wyplaszcz(lista):
    kopia = list(lista)
    for i in range(len(lista)-1,-1,-1):
        del lista[i]
    for x in kopia:
        if type(x)==type([]):
            for y in wyplaszcz(x):
                lista.append(y)
        else:
            lista.append(x)
    return lista
```

```python
l = [4, 1, [ 2, [7] ], 3]
wyplaszcz(l)
l
```
